```python
import pandas as pd
import numpy as np
import seaborn as sns
import altair as alt
```

```python
n=365
dfx = pd.DataFrame(data = {"A":np.random.randn(n), "B":np.random.randn(n)+1},
                  index=pd.date_range(start="2017-01-01", periods=n, freq="D"))
```

```python
dfx
```

```python
alt.Chart(
    dfx.reset_index().melt(id_vars = ["index"], value_vars=["A", "B"]).assign(month = lambda x: x["index"].dt.month)
).mark_boxplot(
    extent='min-max'
).encode(
    alt.X('variable:N', title=''),
    alt.Y('value:Q'),
    column='month:N',
    color='variable:N'
)
```
