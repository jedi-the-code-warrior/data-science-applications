```python
import re   #regular expression

def filter_and_tokenize(raw_text):
    
    cleaned_tokenized_text  = re.findall('\w+', raw_text)
    
    #cleaned_tokenized_text = "".join([char.lower() for char in raw_text if char in filter_])
    
    return cleaned_tokenized_text


letters_numbers_space = string.ascii_letters + string.digits + " "
filter_ = '[%s]+' % re.escape(letters_numbers_space)
filtered_text = df_new['Raw_Text'].apply(lambda x: filter_and_tokenize(x.lower()))

df_new['Cleaned_and_Tokenized'] = filtered_text
#df_new=df_new.drop(['Cleaned_Text'],axis=1)
df_new.head()   
```

```python
from nltk import PorterStemmer 
from nltk.corpus import stopwords

ps = PorterStemmer()
stop_words = stopwords.words('english')

def remove_stop_words(tokenized_text):
    
    no_stop_words = [word for word in tokenized_text if word not in stop_words]
    
    return no_stop_words

df_new['No_Stop_Words'] = df_new['Cleaned_and_Tokenized'].apply(lambda x: remove_stop_words(x))
 

def stemming(tokenized_text):
    
    stemmed_text = [ps.stem(word) for word in tokenized_text]
    
    return stemmed_text

df_new['Stemmed_Text'] = df_new['No_Stop_Words'].apply(lambda x: stemming(x))

df_new.head()
```
