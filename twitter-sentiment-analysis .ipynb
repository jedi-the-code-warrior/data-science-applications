```python
import re   #regular expression

def filter_and_tokenize(raw_text):
    
    cleaned_tokenized_text  = re.findall('\w+', raw_text)
    
    #cleaned_tokenized_text = "".join([char.lower() for char in raw_text if char in filter_])
    
    return cleaned_tokenized_text


letters_numbers_space = string.ascii_letters + string.digits + " "
filter_ = '[%s]+' % re.escape(letters_numbers_space)
filtered_text = df_new['Raw_Text'].apply(lambda x: filter_and_tokenize(x.lower()))

df_new['Cleaned_and_Tokenized'] = filtered_text
#df_new=df_new.drop(['Cleaned_Text'],axis=1)
df_new.head()   
```

```python
from nltk import PorterStemmer 
from nltk.corpus import stopwords

ps = PorterStemmer()
stop_words = stopwords.words('english')

def remove_stop_words(tokenized_text):
    
    no_stop_words = [word for word in tokenized_text if word not in stop_words]
    
    return no_stop_words

df_new['No_Stop_Words'] = df_new['Cleaned_and_Tokenized'].apply(lambda x: remove_stop_words(x))
 

def stemming(tokenized_text):
    
    stemmed_text = [ps.stem(word) for word in tokenized_text]
    
    return stemmed_text

df_new['Stemmed_Text'] = df_new['No_Stop_Words'].apply(lambda x: stemming(x))

df_new.head()
```

```python
df_new['Sentiment_Value'] = np.random.choice([-1,0,1],size=15000, replace=True)
df_new.head()
```

```python
df_final = df_new[['Timestamp', 'Screen_Name','Stemmed_Text','Sentiment_Value']].copy()
df_final.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=stemming)
X_tfidf = tfidf_vect.fit_transform(df_final['Stemmed_Text'])
print(X_tfidf.shape)
print(tfidf_vect.get_feature_names())
```

```python
sample_20 = df_final.sample(20)
sample_20
```

```python
tfidf_vect_sample = TfidfVectorizer(analyzer=stemming)
X_tfidf_sample_20 = tfidf_vect_sample.fit_transform(sample_20['Stemmed_Text'])
print(X_tfidf_sample_20.shape)
print(tfidf_vect_sample.get_feature_names())
```

```python
X_tfidf_df = pd.DataFrame(X_tfidf_sample_20.toarray())
X_tfidf_df.columns = tfidf_vect_sample.get_feature_names()
X_tfidf_df.head()
```

```python
X_features = pd.concat([df_final['Sentiment_Value'], pd.DataFrame(X_tfidf.toarray())], axis=1)
X_features.head()
```

```python
X_features.shape
```

```python
y_label = df_final['Sentiment_Value'].values
y_label.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.model_selection import train_test_split

#print(RandomForestClassifier())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, y_label, test_size=0.2)
```

```python
rfc = RandomForestClassifier(n_estimators=50, max_depth=20, n_jobs=-1)

rfc_fitted_model = rfc.fit(X_train, y_train)

sorted(zip(rfc_fitted_model.feature_importances_, X_train.columns), reverse=True)[0:10]
```

```python
# magic!!!
y_preds = rfc_fitted_model.predict(X_test)
```

```python
precision, recall, fscore, support = score(y_test, y_preds, pos_label=None, average=None)
print('Precision: {} \nRecall: {} \nAccuracy: {}'.format(np.round(precision, 3),
                                                         np.round(recall, 3),
                                                         np.round((y_preds == y_test).sum() / len(y_preds),3)))
```
